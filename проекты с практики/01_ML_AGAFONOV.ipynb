## array 

```python
print(np.zeros((2,5)))
print(np.ones((6,3), dtype="i"))
print(np.linspace(3,14,11))
print(np.random.rand(3,4))
```

```python
list1 = [3,6,2,4]
np_list1 = np.array(list1)

print(list1)
print(np_list1)

print(list1*3)
print(np_list1*3)

list2 = [[1,2,3],[4,5,6]]
np_list2 = np.array(list2)

print(list2)
print(np_list2)

print(np_list2.shape)
print(np_list2.reshape(3,2))
print(np_list2.reshape(1,6))
print(np_list1.shape)
```

```python
A = np.random.rand(2,4)
B = np.ones((4,3))
C = 2*B

B[1,1] = 2

print(A)
print(B)
print(C)
```

```python
print(B+C)
print(B-C)
print(B*C)
print(np.dot(A,B))
print(np.dot(B.T,C))
```

# pygame

```python
import pygame

(width, height) = (640, 280) # размеры окна
background_color = (255,0,255) # цвет фона
screen = pygame.display.set_mode((width, height))
pygame.display.set_caption('Окно Pygame') # заголовок

# Цикл отображения окна
running = True
while running:
    # Обработка пользовательских событий
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
            
    screen.fill(background_color)
    pygame.display.flip()

pygame.quit()
```
