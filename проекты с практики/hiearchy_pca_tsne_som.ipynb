# Иерархическая кластеризация

```python
# Создание полотна для рисования
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(15, 30))
fig.patch.set_facecolor('white')

# Загрузка набора данных "Ирисы Фишера"
from sklearn import datasets
iris = datasets.load_iris()

# Реализация иерархической кластеризации при помощи функции linkage
from scipy.cluster.hierarchy import linkage, dendrogram
mergings = linkage(iris.data, method='ward')

# Построение дендрограммы. Разными цветами выделены автоматически определенные кластеры
R = dendrogram(mergings, labels=[iris.target_names[i] for i in iris.target], orientation = 'left', leaf_font_size = 12)

# Отображение дендрограммы
plt.show()
```

# PCA

```python
# Загрузка набора данных "Ирисы Фишера"
from sklearn import datasets
iris = datasets.load_iris()

from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X = pca.fit_transform(iris['data'])
y = iris['target']

plt.figure(figsize=(13, 10))
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.colorbar()
plt.show()
```

# t-SNE

```python
# Загрузка набора данных "Ирисы Фишера"
from sklearn import datasets
iris = datasets.load_iris()

from sklearn.manifold import TSNE
tsne = TSNE(n_components=2)
X = tsne.fit_transform(iris['data'])
y = iris['target']

plt.figure(figsize = (13, 10))
plt.scatter(X[:, 0], X[:, 1], c = y)
plt.colorbar()
plt.show()
```

# SOM

https://habr.com/ru/post/338868/

Алгоритм обучения SOM:
Вход: $X$, $\eta$, $\sigma$, $\gamma$, $\alpha$, $E$

I. Инициализировать $W$

II. $E$ раз:

1)Перемешать список индексов случайным образом order: = shuffle($[0, 1, \dots, N-2, N-1]$)

2) Для каждого $idx$ из order:

1. $\vec{x} := X[idx]$ // Берём случайный вектор из X

2. $i := \arg \min_j{\left\|\vec{x} - \vec{w_j}\right\|}$ // Находим номер ближайшего к нему нейрона

3. $\forall j: h_{ij} := e^{D_{ij} / \sigma^2}$ // Находим коэффициент близости нейронов по сетке. заметьте, что $h_{ii}$ всегда 1

4. $\forall j: \vec{w_j} := \vec{w_j} + \eta h_{ij} (\vec{x} - \vec{w_j}) $

3) $\eta$ := $\gamma$$\eta$

4) $\sigma$ := $\alpha$$\sigma$

```python
import numpy as np
import matplotlib.pyplot as plt


class node():
    
    def __init__(self, pos):
        self.pos = pos
        self.weights = 0
        
        
class SOM():
    
    def __init__(self, map_size=[20,20]):        
        self.map_size = map_size
        self.sigma = 5
        self.eta = 1.
        self.nodes = []
        for x in range(self.map_size[0]):
            for y in range(self.map_size[1]):
                self.nodes.append(node([x+1/4*(1-(-1)**y), y*np.sqrt(3)/2.]))
        self.dist = np.array([[self.get_dist(node1.pos,node2.pos) \
                               for node2 in self.nodes] for node1 in self.nodes])
        self.fitted = False
        
    def get_dist(self, list1, list2):
        return np.sqrt(sum((i-j)**2 for i,j in zip(list1,list2)))
    
    def fit(self, dataset):
        smpl_size = len(dataset[0])
        # initialization
        for node in self.nodes:
            node.weights = np.random.rand(smpl_size)
        #X_train = np.random.permutation(dataset)
        X_train = dataset[np.random.choice(dataset.shape[0], size=2000, replace=True)]
        sigma = self.sigma
        eta = self.eta
        for data in X_train:            
            winner_indx = np.array([self.get_dist(node.weights,data) for node in self.nodes]).argmin()
            delta = data-self.nodes[winner_indx].weights
            h = np.exp(-self.dist[winner_indx]/sigma)
            for k,node in enumerate(self.nodes):
                node.weights += eta*h[k]*delta
            
            sigma *= .99
            eta *= .99
    
    
from sklearn import datasets
iris = datasets.load_iris()
X_train = iris.data

#X_train = np.array([[np.random.random(),np.random.random(),np.random.random()] for i in range(1000)])

iris_som = SOM(map_size=[20,20])
iris_som.fit(X_train)
#print(np.array([list(iris_som.nodes[i].weights) for i in range(10)]).max())

plt.figure(figsize = (5,5))
X, Y, C = [], [], []
for node in iris_som.nodes:
    X.append(node.pos[0])
    Y.append(node.pos[1])
    if np.min(node.weights) < 0: 
        node.weights -= np.min(node.weights)
    if np.max(node.weights) > 1: 
        node.weights = node.weights/(np.max(node.weights)+.1)
    C.append(list(node.weights[:3]))
plt.scatter(X,Y,c=C,s=100)
plt.show()
```
